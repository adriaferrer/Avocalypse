```python
import pandas as pd
import seaborn as sns
```

```python
kaggle_data = pd.read_csv('/Users/adriaferrer/Desktop/avocado.csv')
```

```python
kaggle_data = kaggle_data.reset_index(drop=True).drop(columns=['Unnamed: 0','4046','4225','4770'])
```

```python
price_per_state = kaggle_data.groupby('region').mean().reset_index()[['region','AveragePrice','Total Volume']]
```

```python
price_per_state.to_csv('price_state.csv')
```

```python
kaggle_data['Date'] = kaggle_data['Date'].astype('datetime64')
```

```python
kaggle_data['Year']=kaggle_data['Date'].str[:4]
kaggle_data['Month'] = kaggle_data['Date'].str[5:7]
kaggle_data['Day'] = kaggle_data['Date'].str[-2:]
```

```python
sns.lineplot(data=kaggle_data,x='Date',y='AveragePrice')
```

```python
kaggle_data_rolling_mean = kaggle_data['AveragePrice'].rolling(24).mean()
```

```python
kaggle_data['rolling_mean'] = kaggle_data_rolling_mean
```

```python
sns.lineplot(data=kaggle_data,x='Date',y='rolling_mean')
```
